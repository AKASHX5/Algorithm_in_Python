```python
fibonachi_cache={}

def fibonachi(n):
    if n in fibonachi_cache:
        return fibonachi_cache[n]
    if n==1:
        value=1
    elif n==2:
        value=1
    elif n>2:
        value=fibonachi(n-1)+fibonachi(n-2)
        
    fibonachi_cache[n]=value
    return value

for n in range(1,101):
    print (n,":",fibonachi(n))

```

```python
from functools import lru_cache

@lru_cache(maxsize=1000)
def fibonachi(n):
    if type(n) != int:
        raise TypeError("n must be integer")
    elif n<1:
        raise ValueError("n must be positive integer")
    elif n==1:
        return 1
    elif n==2:
        return 2
    elif n>2:
        return fibonachi(n-1)+fibonachi(n-2)
    
for n in range(1,11):
    print(n,':',fibonachi(n))
```
